```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = np.linspace(0, 10, 100)
y = np.sin(x)

output_notebook()

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

np.random.seed(42)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.rand(n) * 20
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

output_notebook()

p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

#fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
#counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, LabelSet

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

output_notebook()

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts',
           height=400, toolbar_location=None)

p.vbar(x='fruits', top='counts', source=source, width=0.9, color='skyblue')

labels = LabelSet(x='fruits', y='counts', text='counts', source=source,
                  y_offset=5, text_align='center')

p.add_layout(labels)

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.

#data_hist = np.random.randn(1000)
#hist, edges = np.histogram(data_hist, bins=30)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

output_notebook()

p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           height=400,
           width=600)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='skyblue', line_color='black')

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.

#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource, ColorBar
from bokeh.layouts import column
from bokeh.palettes import Viridis256

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
data_flat = data_heatmap.flatten()

source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

mapper = linear_cmap('value', palette=Viridis256, low=data_flat.min(), high=data_flat.max())

p = figure(title="Heatmap", x_axis_label='X-axis', y_axis_label='Y-axis',
           tools="", width=400, height=400)

p.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color=mapper, line_color=None)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

output_notebook()
show(p)
```
