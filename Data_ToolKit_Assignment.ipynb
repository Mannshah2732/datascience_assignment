## 1. Differences between np.array, np.asarray, and np.asanyarray

* np.array :

  * Creates a new NumPy array from any object (like a list, tuple, etc.).

  * Always returns a new array, even if the input is already an array.
  
  * Allows you to specify the desired data type using the dtype argument.

* np.asarray :

  * Converts the input to an array, but if the input is already an array, it returns the original array (no new copy is made).
  
  * This is more memory-efficient if you know the input is already an array and don’t want to create a new one.

* np.asanyarray :

  * Similar to np.asarray, but it allows for subclasses of ndarray (like masked arrays or other array-like structures) to be returned as their original type.

  * If the input is already an array, it returns it as-is, but if it’s a subclass, it preserves that subclass.

* Example :



```python
import numpy as np
arr1 = np.array([1, 2, 3])

arr2 = np.asarray(arr1)  # arr2 will refer to the same object as arr1 if arr1 is an array.

arr3 = np.asanyarray(arr1)  # Preserves any subclass if the input is one.

print(arr1)
print(arr2)
print(arr3)
```

## 2. Differences between Deep Copy and Shallow Copy

* Shallow Copy :

  * A shallow copy creates a new object, but it does not create copies of nested objects. Instead, it copies references to them.

  * Changes made to the nested objects in the shallow copy will reflect in the original object, as they both refer to the same nested objects.

  * In Python, you can create a shallow copy using the copy module’s copy() function or the list.copy() method.

* Example :

```python
import copy

original = [[1, 2, 3], [4, 5, 6]]
shallowed_copy = copy.copy(original)
shallowed_copy[0][0] = 'x'

print(original)
print(shallowed_copy)
```

* Deep Copy :

  * A deep copy creates a new object and recursively copies all nested objects, so that the new object is completely independent of the original.

  * Changes made to the deep copy do not affect the original object.
  
  * In Python, you can create a deep copy using the copy module’s deepcopy() function.

* Example :

```python
import copy

original = [[1, 2, 3], [4, 5, 6]]
deepied_copy = copy.deepcopy(original)
deepied_copy[0][0] = 'x'

print(original)
print(deepied_copy)
```

```python
# 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

data = pd.DataFrame({'x' : x, 'y' : y})

def get_quadrant(row):
  if row['x'] >= 0 and row['y'] >= 0:
    return 'Quadrant 1'
  elif row['x'] < 0 and row['y'] >= 0:
    return 'Quadrant 2'
  elif row['x'] < 0 and row['y'] < 0:
    return 'Quadrant 3'
  else:
    return 'Quadrant 4'

data['quadrant'] = data.apply(get_quadrant, axis = 1)


plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='quadrant', palette='Set1', s=100)

plt.legend(title='Quadrants')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

plt.show()
```

```python
# 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

from math import pi, sin
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = np.linspace(0, 2 * pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='six(X)', width=800, height=400)

p.grid.grid_line_alpha = 0.5

p.line(x, y, legend_label='sin(x)', line_width=2)

show(p)
```

```python
# 19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import random


output_notebook()

categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = [random.randint(10, 100) for _ in categories]


source = ColumnDataSource(data=dict(categories=categories, values=values))

color_mapper = linear_cmap('values', palette=Viridis256, low=min(values), high=max(values))

p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           width=800, height=400)

p.vbar(x='categories', top='values', source=source, width=0.9, color=color_mapper)

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

show(p)
```
