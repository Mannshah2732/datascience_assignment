10. What are the key assumptions of Multiple Linear Regression

* Multiple Linear Regression (MLR) has several key assumptions that must be met to ensure the model is valid, interpretable, and produces reliable results.

Key Assumptions of Multiple Linear Regression :

1. Linearity

    * The relationship between each independent variable and the dependent variable is linear.
    * Scatter plots or partial regression plots.

2. Independence of Errors (No Autocorrelation)

    * The residuals (errors) should be independent of each other.
    * Especially critical in time-series data.
    * Durbin-Watson test.

3.  Homoscedasticity (Constant Variance of Errors)

    * The variance of the residuals is constant across all levels of the independent variables.
    * Plot residuals vs. predicted values.

4. No Multicollinearity

    * Independent variables should not be highly correlated with each other.
    * High correlation distorts the estimation of coefficients.
    * Variance Inflation Factor (VIF) ‚Äî VIF > 5 or 10 is problematic.

11.  What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model

* Heteroscedasticity occurs when the variance of the residuals (errors) in a regression model is not constant across all levels of the independent variables.

* In Simple Terms :

  In a well-behaved regression model, the spread of residuals should be roughly the same regardless of the value of the predictors ‚Äî this is called homoscedasticity.

  When the spread changes (e.g., gets wider or narrower as ùëã increases), we have heteroscedasticity.

* Example :

  In a residuals vs. predicted values plot:

  Homoscedasticity: Residuals look like a horizontal cloud (equal spread).

  Heteroscedasticity: Residuals fan out or funnel in (triangle or cone shape).

* How Heteroscedasticity Affects the Model :

  1. Incorrect Standard Errors

   * Biased confidence intervals
   * Misleading p-values
   * Potentially wrong conclusions about statistical significance

  2. Coefficient Estimates Remain Unbiased

   * They may still be correct "on average"
   * But they're less precise

  3. Over- or underestimating predictive power

   * The model may seem better (or worse) than it actually is for certain values of predictors.

    * Elastic Net

      * Combines Ridge and Lasso

  Sometimes, multicollinearity arises from small sample sizes. More data can reduce it.



13. What are some common techniques for transforming categorical variables for use in regression models

* Transforming categorical variables into a numerical format is essential before using them in regression models, which require numerical inputs.

Here are the most common techniques :

1. One-Hot Encoding (Dummy Variables)

* Converts each category into a separate binary column (0 or 1).

* Commonly used for nominal (unordered) categories.

2. Label Encoding

* Assigns each category a unique number.

* Best for ordinal variables where order matters (e.g., Low, Medium, High).

3. Ordinal Encoding (Custom Mapping)

* Map categories to numbers based on meaningful order.

4. Target (Mean) Encoding

* Replace each category with the mean of the target variable for that category.

* Useful for high-cardinality features (e.g., many city names).

5. Frequency Encoding

* Replace each category with its frequency (count) in the dataset.

6. Binary Encoding / Hashing (Advanced)

* Useful for high-cardinality features

* More compact than one-hot encoding

18.  What are the limitations of using R¬≤ as a sole measure of model performance.

* The coefficient of determination (R¬≤) measures how well your independent variables explain the variability of the dependent variable.

  While it‚Äôs useful, relying on R¬≤ alone can be misleading.

  Ranges from 0 to 1 (or 0% to 100%)

  Higher R¬≤ ‚áí more variance in the target variable is explained by the model

* Why You Shouldn‚Äôt Rely Solely on R¬≤:

1. Does Not Indicate Model Accuracy

  * A high R¬≤ doesn‚Äôt mean predictions are close to actual values.

  * It measures fit, not error.

2. Does Not Detect Overfitting

  * R¬≤ always increases when you add more variables ‚Äî even if they're irrelevant.

  * This can give a false sense of improvement.

3. Not Comparable Across Different Datasets

  * R¬≤ is dataset-specific.

  * You can‚Äôt compare R¬≤ from different models on different datasets.

4.  Assumes Linearity

  * R¬≤ assumes a linear relationship.

  * If your data has non-linear trends, R¬≤ may be low even for a good model.

5.  Insensitive to Bias

  * R¬≤ doesn‚Äôt reveal if your model consistently overpredicts or underpredicts.

28. What are the limitations of polynomial regression.

While polynomial regression is a powerful tool for modeling non-linear relationships, it comes with several important limitations that should be carefully considered :

1. Overfitting (High Variance)

  * As the degree of the polynomial increases, the model becomes more flexible ‚Äî but this can lead to overfitting.

  * The model may fit the training data too closely, capturing noise instead of the actual trend.

2. Extrapolation is Unreliable

  * Polynomial regression is very unstable outside the range of the training data.

  * Predictions for ùë• values far beyond the training range can be wildly inaccurate, especially for higher degrees.

3. Multicollinearity

  * Adding polynomial and interaction terms can introduce high correlation among features (e.g., ùë•, ùë•2, ùë•3).

  * This makes coefficient estimates unstable and hard to interpret.

4. Increased Complexity

  * More terms mean more parameters to estimate, increasing computational cost.

  * High-degree models can be hard to interpret, especially with multiple variables.

5. Sensitive to Outliers

  * Outliers can have a disproportionate influence on the fitted curve, especially in high-degree models.

6. Harder to Select the Right Degree

  * There is no fixed rule for choosing the best polynomial degree.

  * Requires model selection techniques like:

    * Cross-validation

    * Adjusted R¬≤

    * AIC/BIC

  * Unlike regular R¬≤, this metric penalizes model complexity.

  * It increases only if adding a new term improves the model more than expected by chance.

3. AIC / BIC (Akaike/Bayesian Information Criterion)

4. Residual Plots

  * Check for patterns in residuals:

30. Why is visualization important in polynomial regression.

* Visualization plays a critical role in polynomial regression because it helps you understand, evaluate, and communicate how well your model is capturing the underlying patterns in the data ‚Äî especially when those patterns are non-linear.

* Key Reasons Why Visualization Matters :

1. Detecting Underfitting or Overfitting

  * A linear fit may miss curves (underfitting).

  * A high-degree polynomial might wiggle too much (overfitting).

2. Understanding the Model‚Äôs Behavior

  * Polynomial regression introduces curvature in predictions.

  * Plots help you interpret :

    * Where the curve bends

    * Whether the trend makes sense

    * How extreme predictions behave (especially at the edges)

3. Checking for Extrapolation Risks

  * Polynomial models often behave unpredictably outside the data range.

  * Visuals help spot unrealistic spikes or dips at the edges of the curve.

4. Validating Residuals and Errors

  * Residual plots help check whether errors are :

    * Randomly distributed (good)

    * Showing patterns (bad ‚Äî possibly underfitting or misspecification)

5. Choosing the Right Degree

  * Plotting models of different degrees side-by-side makes it easy to :

    * Spot the best-fitting curve

    * Avoid unnecessary complexity

     X_test = np.linspace(0, 8, 100).reshape(-1, 1)
